```python
# Trabalhando com listas em Python

matriz = [
    [1, "a", 2]
    ["b", 3, 4]
    [6, 5, "c"]
]

matriz[0] #[1, "a", 2]
matriz[0][0] # 1
matriz[0] [-1] # 2
matriz [-1] [-1] # "c"

# Fatiamento

lista = ["P", "y", "t", "h", "o", "n"]

lista[2:] # ["t", "h", "o", "n"]
lista[:2] # ["P", "y"]
lista[1:3] # ["y", "t"]
lista[0:3:2] # ["P", "t"]
lista[::] # ["P", "y", "t", "h", "o", "n"]
lista[::-1] # ["n", "o", "h", "t", "y", "P"]
```

```python
""" Compreensão de listas- oferece sintaxe mais curta para
criar uma nova lista com base nos valores de uma lista
existente(filtro) ou gerar uma nova lista aplicando alguma
modificação nos elementos de uma lista existente

"""
numeros = [1, 30, 21, 2, 9, 65, 34]
pares = []

for numero in numeros:
  if numero % 2 == 0:
      pares.append(numero)
```

```python
# modificar os numeros da lista ao quadrado

numeros = [1, 30, 21, 2, 9, 65, 34]
quadrado = []

for numero in numeros:
  quadrado.append(numero ** 2)

print(quadrado)

# outra forma

numeros = [1, 30, 21, 2, 9, 65, 34]
quadrado = [numero ** 2 for numero in numeros]

print(quadrado)
```

```python
# métodos da classe list

# append()- adiciona um elemento à lista

lista = []

lista.append(1)
lista.append("Python")
lista.append([40, 30, 20])

print(lista)

# clear()- limpa os elementos da lista

lista = [1, 'Python', [40, 30, 20]]

lista.clear()
print(lista)

#copy()- imprime a mesma lista porém com uma instância diferente
lista = [1, 'Python', [40, 30, 20]]

l2 = lista.copy()

print(lista)

print(id(l2), id(lista))

l2[0] = 2

print(l2)
print(lista)


```

```python
# count()- mostra o numeros de vezes que um elemento aparece na lista

cores = ["vermelho", "azul", "verde", "azul"]

print(cores.count("vermelho"))
print(cores.count("azul"))
print(cores.count("verde"))

# extend()- adicionar uma outra lista(juntar outra lista)

linguagens = ["python", "js", "c"]

print(linguagens)

linguagens.extend(["java", "c"])

print(linguagens)
```

```python
# index()- mostra qual é a primeira ocorrência do objeto

linguagens = ["python", "js", "c", "java", "csharp"]

print(linguagens.index("java"))
print(linguagens.index("python"))

linguagens = ["python", "java", "c", "java", "csharp"]
print(linguagens.index("java"))

# pop()- remove o ultimo elemento da lista
# é possível passar qual índice deseja remover Ex.

linguagens = ["python", "js", "c", "java", "csharp"]

print(linguagens.pop())
print(linguagens.pop())
print(linguagens.pop())
print(linguagens.pop(0))
```

```python
# index()- mostra qual é a primeira ocorrência do objeto

linguagens = ["python", "js", "c", "java", "csharp"]

print(linguagens.index("java"))
print(linguagens.index("python"))

linguagens = ["python", "java", "c", "java", "csharp"]
print(linguagens.index("java"))

# pop()- remove o ultimo elemento da lista
# é possível passar qual índice deseja remover Ex: print(linguagens.pop(0))

linguagens = ["python", "js", "c", "java", "csharp"]

print(linguagens.pop())
print(linguagens.pop())
print(linguagens.pop())
print(linguagens.pop(0))

# remove()- outra forma de remover objetos da lista
# diferente do pop(), ao invés de passar o índice, passa-se o nome do objeto

linguagens = ["python", "js", "c", "java", "csharp"]

linguagens.remove("c")

print(linguagens)
```

```python
# reverse() inverter a lista

linguagens = ["python", "js", "c", "java", "csharp"]

linguagens.reverse()

print(linguagens)
```

```python
# sort() coloca a lista em ordem

linguagens = ["python", "js", "c", "java", "csharp"]

linguagens.sort() #lista ordenada
print(linguagens)
linguagens.sort(reverse=True) #lista ordenada reversa
print(linguagens)
linguagens.sort(key=lambda x:len(x)) # lista em ordem crescente por tamanho da palavra
print(linguagens)
linguagens.sort(key=lambda x:len(x), reverse=True) #lista em ordem decrescente por tamanho
print(linguagens)
```
